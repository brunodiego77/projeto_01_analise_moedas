```python
import matplotlib.pyplot as plt

# 1. Cria a tabela correta com as máximas reais (sem a armadilha do Pandas)
df_maximos = df.sort_values(by="high", ascending=False).drop_duplicates(subset=["name"])

# 2. Ordena do menor para o maior preço para o gráfico ficar bonito
df_maximos = df_maximos.sort_values(by="high", ascending=True)

# 3. Criação do Gráfico de Barras Horizontais
plt.figure(figsize=(12, 7))

# Desenha as barras
plt.barh(df_maximos["name"], df_maximos["high"], color="royalblue", edgecolor="black")

# O TRUQUE DE MESTRE: Transforma o eixo X em escala Logarítmica
plt.xscale("log") 

plt.xlabel("Valor Máximo em Reais (Escala Logarítmica)", fontsize=12, fontweight="bold")
plt.ylabel("Moeda", fontsize=12)
plt.title("Pico Histórico das Moedas e Cripto (2025-2026)", fontsize=16, fontweight="bold")

# Adiciona uma grade vertical para facilitar a leitura
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()

# Mostra o gráfico
plt.show()
```
