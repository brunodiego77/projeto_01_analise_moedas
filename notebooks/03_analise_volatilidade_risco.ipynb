```python
import pandas as pd

# 1. Carregando os dados
df_moedas = pd.read_csv(r"C:\Users\Robert\Desktop\projeto_01\data\stg\stg_moedas_silver.csv")
df_comm = pd.read_csv(r"C:\Users\Robert\Desktop\projeto_01\data\stg\stg_commodities_silver.csv") # O Ouro mora aqui!



# 2. Convertendo as datas para garantir a ordem
df_moedas["data"] = pd.to_datetime(df_moedas["data"])
df_comm["data"] = pd.to_datetime(df_comm["data"])

# 3. Isolando os nossos 3 protagonistas
df_btc = df_moedas[df_moedas["code"] == "BTC"].sort_values("data")
df_usd = df_moedas[df_moedas["code"] == "USD"].sort_values("data")
df_ouro = df_comm[df_comm["code"] == "XAU"].sort_values("data") # XAU é o código universal do Ouro

# 4. Calculando a variação percentual diária (O quanto o preço mudou de um dia pro outro)
# Usamos o .pct_change() para ver os "sustos" diários e o .std() para tirar a média da volatilidade
volatilidade_btc = df_btc["bid"].pct_change().std() * 100
volatilidade_usd = df_usd["bid"].pct_change().std() * 100
volatilidade_ouro = df_ouro["bid"].pct_change().std() * 100

print("🏆 RANKING DE VOLATILIDADE (RISCO) 🏆")
print("-" * 40)
print(f"1. Bitcoin (Ativo de Risco): {volatilidade_btc:.2f}% de variação média")
print(f"2. Ouro (Reserva Física): {volatilidade_ouro:.2f}% de variação média")
print(f"3. Dólar (Reserva Fiduciária): {volatilidade_usd:.2f}% de variação média")
```
